```python
#  Stack
#  Last in and First Out
# back button in chrome , reverse a word

# creating a stack

def create_stack():
    stack = []
    return stack

#  check the stack is empty or not 

def checking_stack(stack):
    if len(stack)==0:
        return stack

# Adding the items into stack
# 
def adding_stack(stack,x):
    stack.append(x)
    print(stack)

# removing the elements from the stack 

def pop(stack):
    if checking_stack(stack):
        return "stack is empty"
    return stack.pop()

stack = create_stack()
adding_stack(stack, str(1))
adding_stack(stack, str(2))
adding_stack(stack, str(3))

pop(stack)
print(stack)
```

```python
# Queue Data Structure 
# First in First Out

# - Enqueue  -- adding element to the queue
# - Dequeue  -- Removing an element from the front 
# - IsFull   -- queue is full
# - Is Empty -- queue is empty 
# - peek  -- getting the value of the front without removing 


# example 
#  -    call center phone calls 
#  -    tickets in railway stations

```

```python
# example of creating a queue

class Queue:
    def __init__(self):
        self.queue = []
    def adding(self, item):
        self.queue.append(item)
    def removing(self):
        self.queue.pop(0)
    
    # dispaly the size 
    def display(self):
        print(self.queue)
    def size(self):
        print(len(self.queue))


q = Queue()
q.adding(1)
q.adding(2)
q.adding(3)
q.adding(4)

q.display()
q.size()

q.removing()

q.display()
q.size()
```

```python
# Double ended Queue 

#  - Input Restricted Deque  - input is from one end only but we can delete from two ends
#  - Output Restricted Deque  - output is from one end only but we can input from two ends 


```

```python
# example for the doble ended queue

class DoubleQueue:
    def __init__(self):
        self.item =[]
    def to_remove_from_starting(self):
        self.item.pop(0)
    def to_remove_from_end(self):
        self.item.pop()
    
    def addFront(self, x):
        self.item.insert(0,x)
    def addEnd(self,y):
        self.item.append(y)

    def size(self):
        print(len(self.item))

    def isEmpty(self):
        return self.item ==[]
    
```

```python
d = DoubleQueue()
print(d.isEmpty())
d.addEnd(6)
d.addEnd(2)
d.addFront(4)
d.addFront(9)
d.size()
print(d.item)
print(d.to_remove_from_end())
print(d.to_remove_from_starting())

print(d.size())
print(d.size)
```
